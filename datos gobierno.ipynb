```python
def yesno (column_name,dataset,key,value):
    if len(key)==len(value):
        dict_new=zip(key,value)
        key_values=dict(dict_new)
        for i in column_name:
            dataset[i] = dataset[i].map(key_values)
        return dataset 

    else:
        return 'The len of the key and value must be the same'
          
```

```python
def Dias_contagio (test):
    test["DATE_DIFF"] = test["FECHA_INGRESO"] - test["FECHA_SINTOMAS"]
    test.DATE_DIFF.dt.days
    g=test.DATE_DIFF.astype(int)/86400000000000
    G=pd.Series.to_frame(g,name='DIAS_TRANSCURRIDOS')
    COVID19_contagios = test.merge(G, left_index=True, right_index=True)
    COVID19_contagios=COVID19_contagios.drop(['DATE_DIFF'], axis=1)
    return COVID19_contagios
```
